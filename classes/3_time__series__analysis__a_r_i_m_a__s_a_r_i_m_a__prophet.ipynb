# Time-Series Analysis: Unlocking Insights in Time-Dependent Data

Suppose you're a meteorologist tasked with forecasting next month‚Äôs rainfall based on years of past data. How do you analyze the patterns and predict future values? This is where time-series analysis comes in‚Äîa statistical tool to uncover trends, patterns, and relationships in data collected over time, empowering you to make accurate forecasts.


# What Is Time-Series Analysis?

Time-series analysis is a powerful statistical technique used to analyze data points collected or recorded at specific time intervals. This method helps uncover trends, patterns, and relationships in time-dependent data, which can be leveraged for forecasting and decision-making.


Time-series analysis enables us to recognize apparent trends and consistent patterns in a series. It captures the movement of a variable across time. In continuous time-series analysis, the variable is examined continuously, though the series is typically defined at specific points rather than every point in time.


### **Applications**

-**Stock Market Analysis**: Predicting stock prices.

-**Weather Forecasting**: Analyzing temperature or rainfall patterns.

-**Economic Indicators**: Monitoring inflation or GDP growth.

-**Demand Forecasting**: Anticipating product demand.

-**IoT Sensor Data**: Tracking performance metrics over time.

---

## **Assumptions in Time-Series Analysis**

To build a robust time-series model, the following assumptions must be met:

1. **Sufficient Data Points**  
   - At least 50 independent data points for meaningful analysis.

2. **Data Integrity**  
   - Handle missing values and outliers to maintain accuracy.

3. **Stationarity**  
   - The series' statistical properties (mean, variance, autocorrelation) should remain constant over time.

4. **Nonstationary Noise**  
   - Avoid random, meaningless variations.

5. **Autocorrelation**  
   - Recognize relationships between current and past values.


---

## **Types of Time-Series Models**

### 1. **Smoothing and Moving Averages (MA)**  
   - Its is a way to smooth out the ups and downs(noise) in the data by averaging past values. It helps us see the overall trend.  
 See supplementary reading [Time Series Analysis](https://colab.research.google.com/drive/1yn9yC6gsW_Yk1l-V55OoQHHm3LujWQ_0)

### 2. **Autoregressive (AR) Models**  
   - Explains a variable using its own lagged values. that is the model predicts future values based on past values. For example, if temperatures have been rising, it assumes they might keep rising.

### 3. **ARMA Models (AR + MA)**  
   - Combines AR and MA methods for series analysis. It uses both past values and past errors (differences between predicted and actual values) to make predictions.

### 4. **ARIMA Models (Autoregressive Integrated Moving Averages)**  
   - This is an upgraded version of ARMA. It can handle data that has trends or patterns that change over time **(non-stationary data)**. It ‚Äúdetrends‚Äù the data first, then applies ARMA.
   - **Seasonal ARIMA (SARIMAX)**: Incorporates seasonal components and **exogenous variables**, For example, if sales spike every December, SARIMAX can account for that. The ‚ÄúX‚Äù means it can also include external factors, like holidays.

### 5. **ARCH Models (Autoregressive Conditional Heteroscedasticity)**  
   - This is used when the data‚Äôs volatility (how much it jumps up and down) changes over time. It‚Äôs common in financial data like stock prices.

### 6. **VARMA Models (Vector ARMA)**  
   - This is for analyzing multiple time series that affect each other. For example, if you‚Äôre looking at both temperature and ice cream sales, VARMA can model how they influence each other.

## **Steps in Time-Series Analysis**

1. **Visualize the Data**  
   - Plot the series to identify trends, seasonality, and outliers.

2. **Test for Stationarity**  
   - Use statistical tests to confirm whether the series is stationary.

3. **Transform Data if Necessary**  
   - Apply differencing or log transformations for stationarity.

4. **Model Selection**  
   - Choose an appropriate model (ARIMA, SARIMAX, etc.) based on the series' structure.

5. **Model Training and Validation**  
   - Split the data into training and testing sets. Fit the model and evaluate performance.

6. **Forecasting**  
   - Use the model to predict future values.

---

## **Common Pitfalls in Time-Series Analysis**

1. **Overfitting**  
   - Avoid overly complex models that do not generalize well.

2. **Ignoring Assumptions**  
   - Ensure all assumptions are validated before proceeding.

3. **Seasonality Neglect**  
   - Failing to account for seasonal patterns can lead to inaccurate forecasts.





## Predicting USD/NGN Exchange Rate Fluctuations using Advanced Time Series Models

The USD/NGN exchange rate is a critical indicator of Nigeria's economic health. Accurate forecasting of exchange rate fluctuations can inform investment decisions, trade policies, and monetary interventions. This project employs advanced time series models, including ARIMA, SARIMAX, and Prophet, to predict USD/NGN exchange rates and provide insights into the underlying dynamics.

```python
import yfinance as yf
import pandas as pd

# ticker symbol for USD/NGN
ticker_symbol = "NGN=X"

# Fetch historical data for the last 10 years up to 2024
data = yf.download(ticker_symbol, start="2014-01-01", end="2024-12-31")

data.to_csv('USD_NGN_10_years.csv')
```

```python
# Loading data from Google Drive instead of local file
import requests
from io import StringIO

def read_gd(sharingurl):
    file_id = sharingurl.split('/')[-2]
    download_url='https://drive.google.com/uc?export=download&id=' + file_id
    url = requests.get(download_url).text
    csv_raw = StringIO(url)
    return csv_raw

url = "https://drive.google.com/file/d/1VulCAO7Y4oCkJvV8bIc5Tqg5kaI5f5dM/view?usp=sharing"
gdd = read_gd(url)
```

```python
import pandas as pd

# Read the CSV with proper datetime parsing
data = pd.read_csv(
    gdd,
    parse_dates=['Date'],          # Convert 'Date' to datetime
    dayfirst=True                  # Important for DD/MM/YYYY format
)

# Convert numeric columns
numeric_cols = ['Price', 'Open', 'High', 'Low',]
for col in numeric_cols:
    if col in data.columns:
        data[col] = data[col].astype(str).str.replace(',', '').astype(float)

# Rename 'Price' to 'Close'
data = data.rename(columns={'Price': 'Close'})

# Set 'Date' as index (optional, for time-series analysis)
data = data.set_index('Date')

# Save to CSV
# data.to_csv('USD_NGN_formatted.csv', float_format='%.6f')
```

## Visualizing Data

```python
import matplotlib.pyplot as plt

# Visualize the "Close" prices
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='Close Price', color='blue')
plt.title("USD/NGN Exchange Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.show()
```

The plot shows significant trends and possibly structural breaks in the USD/NGN exchange rate over time. There is a clear upward trend with some periods of sharp fluctuations, particularly after 2020.

## Testing for Stationarity

```python
from statsmodels.tsa.stattools import adfuller

# Perform the ADF test
adf_test = adfuller(data['Close'])

# Display the results
adf_results = {
    "Test Statistic": adf_test[0],
    "p-value": adf_test[1],
    "Lags Used": adf_test[2],
    "Number of Observations Used": adf_test[3],
    "Critical Values": adf_test[4],
}
for key, value in adf_results.items():
    print(f"{key}: {value}")
```

The **p-value (0.9893)** is much greater than the *significance threshold (0.05)*, meaning we fail to reject the null hypothesis. The series is not stationary.The Test Statistic (1.074) is higher than all the critical values, further confirming that the series is **not stationary**.

## Differencing the Data

To make the series *stationary*, we will apply **first-order differencing**. This removes trends and makes the series stationary.

```python
# Apply first-order differencing
data['Differenced'] = data['Close'].diff()

# Visualize the differenced series
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Differenced'], label='Differenced Close Price', color='green')
plt.title("Differenced USD/NGN Exchange Rate")
plt.xlabel("Date")
plt.ylabel("Differenced Close Price")
plt.legend()
plt.grid()
plt.show()

# Test for stationarity on the differenced series
adf_test_diff = adfuller(data['Differenced'].dropna())
adf_results_diff = {
    "Test Statistic": adf_test_diff[0],
    "p-value": adf_test_diff[1],
    "Lags Used": adf_test_diff[2],
    "Number of Observations Used": adf_test_diff[3],
    "Critical Values": adf_test_diff[4],
}
for key, value in adf_results_diff.items():
    print(f"{key}: {value}")
```

The differenced series appears more stable, with fluctuations around a mean of zero. This suggests that the trend has been removed, and the series is closer to being stationary.

The Test Statistic (-9.151) is much lower than the critical values, and the **p-value is far below 0.05**.
We can reject the null hypothesis, indicating that the differenced series is now *stationary*.

## Model Selection

### ARIMA

ARIMA (p, d, q) finds a key trend and reasonably predicts future values of the series,
where **p** represents the order of the autoregressive model, **d** represents the degree of
differencing, and **q** represents the order of the moving average model. The term ARIMA
is threefold: *AR (autoregressive)* represents the linear combination of preceding values‚Äô
influence, *I (integrative)* represents the random walk, and the *MA (moving average)*
represents the linear combination of preceding errors.

```python
# Determine the optimal ARIMA order

from pmdarima import auto_arima

auto_arima_model = auto_arima(data['Close'],
                              seasonal=False,
                              trace=True,
                              error_action='ignore',
                              suppress_warnings=True,
                              stepwise=True)

print(auto_arima_model.summary())
```

The auto_arima output suggests that the best model is ARIMA(3, 1, 2) based on the lowest AIC value. This means:


    3 autoregressive terms (AR): the model looks back at 3 previous time points to help predict the current value.

    1 difference (I): it subtracts the previous value to make the data more stable.

    2 moving average terms (MA): it adjusts for noise based on the errors in the previous two predictions.





**AIC (Akaike Information Criterion)** ***(23343.458)*** & **BIC (Bayesian Information Criterion)** ***(23385.190)***: These are scoring rules‚Äîlower values generally mean a better fit

## Train ARIMA Model

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA(3, 1, 2) model
arima_model = ARIMA(data['Close'], order=(3, 1, 2))
arima_result = arima_model.fit() # training

# Print model summary
print(arima_result.summary())

# Plot the fitted values against the actual values
data['Fitted'] = arima_result.fittedvalues

plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Actual', color='blue')
plt.plot(data['Fitted'], label='Fitted', color='red')
plt.title("ARIMA(3, 1, 2) Model - Actual vs Fitted Values")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.show()
```

The red line (fitted values) aligns well with the blue line (actual values), confirming the model captures the series' behavior effectively.

## Forcasting

Now that the ARIMA(3,1,2) model fits well, we can use it for forecasting future USD/NGN exchange rates

```python
# Forecast the next 30 days
forecast_steps = 30
forecast = arima_result.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1], periods=forecast_steps + 1, freq='B')[1:]

# Extract forecasted values and confidence intervals
forecast_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Historical', color='blue')
plt.plot(forecast_index, forecast_values, label='Forecast', color='green')
plt.fill_between(
    forecast_index,
    confidence_intervals.iloc[:, 0],
    confidence_intervals.iloc[:, 1],
    color='lightgreen',
    alpha=0.5,
)
plt.title("USD/NGN Exchange Rate Forecast")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.show()
```

## SARIMAX Model

SARIMAX extends ARIMA by including seasonal components and exogenous variables if needed.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define SARIMAX model parameters (tweak P, D, Q, s as necessary)
seasonal_order = (1, 1, 1, 12)  # Example: Seasonal ARIMA(1,1,1) with a 12-month seasonality
order = (3, 1, 2)  # Based on ARIMA results

# Fit SARIMAX model
sarimax_model = SARIMAX(data['Close'], order=order, seasonal_order=seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
sarimax_result = sarimax_model.fit() # Training
```

```python
# Print model summary
print(sarimax_result.summary())
```

```python
# Forecast the next 30 days
forecast_steps = 30
sarimax_forecast = sarimax_result.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data.index[-1], periods=forecast_steps + 1, freq='B')[1:]
forecast_values = sarimax_forecast.predicted_mean
confidence_intervals = sarimax_forecast.conf_int()
```

```python
# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(data['Close'], label='Historical', color='blue')
plt.plot(forecast_index, forecast_values, label='SARIMAX Forecast', color='green')
plt.fill_between(
    forecast_index,
    confidence_intervals.iloc[:, 0],
    confidence_intervals.iloc[:, 1],
    color='lightgreen',
    alpha=0.5,
)
plt.title("USD/NGN Exchange Rate Forecast (SARIMAX)")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.show()
```

## Prophet Model

Prophet is robust to missing data and handles seasonality automatically. It‚Äôs suitable for business time series with daily/weekly seasonality.

```python
# pip install prophet
```

```python
from prophet import Prophet

# Prepare data for Prophet
prophet_data = data['Close'].reset_index()
prophet_data.columns = ['ds', 'y']

# Initialize and fit the Prophet model
prophet_model = Prophet()
prophet_model.fit(prophet_data)  #training
```

```python
# Forecast the next 30 days
future = prophet_model.make_future_dataframe(periods=30)
forecast = prophet_model.predict(future)

# Plot the forecast
fig = prophet_model.plot(forecast)
plt.title("USD/NGN Exchange Rate Forecast (Prophet)")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid()
plt.show()
```

```python
# Plot components (trend and seasonality)
fig2 = prophet_model.plot_components(forecast)
plt.show()
```

```python
# Get the forecast dataframe
future = prophet_model.make_future_dataframe(periods=30)
forecast = prophet_model.predict(future)

# Select the last 30 days (forecast period) and relevant columns
forecast_table = forecast.tail(30)[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

# Rename columns for clarity
forecast_table = forecast_table.rename(columns={
    'ds': 'Date',
    'yhat': 'Predicted Value',
    'yhat_lower': 'Lower Bound',
    'yhat_upper': 'Upper Bound'
})

# Format numbers to 2 decimal places
forecast_table['Predicted Value'] = forecast_table['Predicted Value'].round(2)
forecast_table['Lower Bound'] = forecast_table['Lower Bound'].round(2)
forecast_table['Upper Bound'] = forecast_table['Upper Bound'].round(2)

# Display the table
print("\nUSD/NGN Exchange Rate 30-Day Forecast:")
print(forecast_table.to_string(index=False))

# Optionally, save to CSV
# forecast_table.to_csv('usd_ngn_forecast.csv', index=False)
```
